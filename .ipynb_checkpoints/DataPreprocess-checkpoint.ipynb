```python
# Cell 2
import os, glob, gc
import pandas as pd
import numpy as np
from datetime import datetime
import joblib
from tqdm import tqdm

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.metrics import classification_report, mean_squared_error, r2_score

# Paths - edit if you moved files
UPLOADED_BTS = "/mnt/data/T_ONTIME_REPORTING-2.csv"
UPLOADED_MAPPING = "/mnt/data/T_ONTIME_REPORTING.csv"  # optional mapping file with id->IATA
OPENFLIGHTS_URL = "https://raw.githubusercontent.com/jpatokal/openflights/master/data/airports.dat"
OPENFLIGHTS_CACHE = "data/openflights_airports.csv"

NORMALIZED_BTS = "data/bts/T_ONTIME_NORMALIZED_sample.csv"
PROCESSED_OUTPUT = "data/processed_flights_sample.parquet"
PREPROCESSOR_PATH = "models/preprocessor_sample.joblib"
CLASSIFIER_PATH = "models/rf_classifier_sample.joblib"
REGRESSOR_PATH = "models/rf_regressor_sample.joblib"

# Memory-conservative defaults
SAMPLE_N = 5000      # reduce to 1000 if kernel still restarts; set to None only if you have enough RAM
DELAY_THRESH = 15    # minutes threshold for on-time
RANDOM_STATE = 42

os.makedirs("data/bts", exist_ok=True)
os.makedirs("data", exist_ok=True)
os.makedirs("models", exist_ok=True)

print("Config: SAMPLE_N =", SAMPLE_N)
```

```python
# Cell 3
def find_bts_path():
    candidates = [
        UPLOADED_BTS,
        os.path.join("data/bts", os.path.basename(UPLOADED_BTS)),
        *sorted(glob.glob("data/bts/*.csv"))
    ]
    for p in candidates:
        if p and os.path.exists(p):
            return p
    raise FileNotFoundError("BTS CSV not found. Place it at " + UPLOADED_BTS + " or data/bts/")

bts_path = find_bts_path()
print("Loading BTS from:", bts_path)

read_kwargs = {"low_memory": False}
if SAMPLE_N:
    read_kwargs["nrows"] = SAMPLE_N

bts_df = pd.read_csv(bts_path, **read_kwargs)
print("Loaded BTS shape:", bts_df.shape)
# quick memory hint
import psutil, os as _os
print("Process memory (GB):", psutil.Process(_os.getpid()).memory_info().rss/1e9)
```

```python
# Cell 4
# Normalize header names
bts_df.columns = [c.strip().upper().replace(" ", "_") for c in bts_df.columns]
print("Columns sample:", bts_df.columns.tolist()[:80])

# detect common columns
col_candidates = {
    'ORIGIN_ID': ['ORIGIN_AIRPORT_ID','ORIGIN_AIRPORT','ORIGIN'],
    'DEST_ID': ['DEST_AIRPORT_ID','DEST_AIRPORT','DEST'],
    'ORIGIN_CITY': ['ORIGIN_CITY_NAME'],
    'DEST_CITY': ['DEST_CITY_NAME'],
    'CRS_DEP': ['CRS_DEP_TIME','CRSDEP','CRS_DEP'],
    'CRS_ARR': ['CRS_ARR_TIME','CRSARR','CRS_ARR'],
    'ARR_DELAY': ['ARR_DELAY','ARRDELAY','ARR_DELAY_NEW'],
    'DEP_DELAY': ['DEP_DELAY','DEPDELAY','DEP_DELAY_NEW'],
    'FL_DATE': ['FL_DATE']
}

detected = {}
for key, opts in col_candidates.items():
    found = [o for o in opts if o in bts_df.columns]
    detected[key] = found[0] if found else None
print("Detected fields:", detected)

# Rename detected to canonical short names if found
rename = {}
for canon, actual in detected.items():
    if actual:
        rename[actual] = canon
bts_df = bts_df.rename(columns=rename)

# parse FL_DATE if present and fill YEAR/MONTH/DAYOFMONTH
if 'FL_DATE' in bts_df.columns:
    bts_df['FL_DATE'] = pd.to_datetime(bts_df['FL_DATE'], errors='coerce')
    if 'YEAR' not in bts_df.columns: bts_df['YEAR'] = bts_df['FL_DATE'].dt.year
    if 'MONTH' not in bts_df.columns: bts_df['MONTH'] = bts_df['FL_DATE'].dt.month
    if 'DAYOFMONTH' not in bts_df.columns: bts_df['DAYOFMONTH'] = bts_df['FL_DATE'].dt.day

# Save normalized small CSV for reproducibility
bts_df.to_csv(NORMALIZED_BTS, index=False)
print("Saved normalized sample to:", NORMALIZED_BTS)
```

```python
# Cell 5
if not os.path.exists(OPENFLIGHTS_CACHE):
    cols = ["airport_id","name","city","country","iata","icao","lat","lon","altitude","timezone","dst","tz_db","type","source"]
    af = pd.read_csv(OPENFLIGHTS_URL, header=None, names=cols)
    af.to_csv(OPENFLIGHTS_CACHE, index=False)
else:
    af = pd.read_csv(OPENFLIGHTS_CACHE, low_memory=False)

# Normalize
af['iata'] = af['iata'].astype(str).str.upper().replace({'NAN':None})
af['city_norm'] = af['city'].astype(str).str.upper().str.replace(r"[^A-Z0-9\s]"," ", regex=True).str.strip()
af['name_norm'] = af['name'].astype(str).str.upper().str.replace(r"[^A-Z0-9\s]"," ", regex=True).str.strip()
print("OpenFlights rows:", len(af))
```

```python
# Cell 6
id_to_iata = {}
if os.path.exists(UPLOADED_MAPPING):
    try:
        map_df = pd.read_csv(UPLOADED_MAPPING, low_memory=False)
        map_df.columns = [c.strip().upper().replace(" ", "_") for c in map_df.columns]
        # heuristics
        id_cols = [c for c in map_df.columns if 'AIRPORT_ID' in c or c.endswith('_ID') or c=='ID']
        iata_cols = [c for c in map_df.columns if c in ['IATA','IATA_CODE','AIRPORT_CODE','CODE','LOCID','LOCATIONID']]
        if id_cols and iata_cols:
            id_col = id_cols[0]; iata_col = iata_cols[0]
            tmp = map_df[[id_col, iata_col]].dropna()
            tmp[iata_col] = tmp[iata_col].astype(str).str.upper()
            try:
                keys = tmp[id_col].astype(int).tolist()
                vals = tmp[iata_col].tolist()
                id_to_iata.update(dict(zip(keys, vals)))
                print("Built id->iata mapping entries:", len(id_to_iata))
            except Exception:
                print("Mapping file columns not convertible to int; skipping direct mapping.")
        else:
            print("Mapping file found but no obvious id/iata columns.")
    except Exception as e:
        print("Could not load mapping file:", e)
else:
    print("No mapping file at", UPLOADED_MAPPING)
```

```python
#cell 7
origin_id_col = None
dest_id_col = None
for c in ['ORIGIN_AIRPORT_ID','ORIGIN_ID','ORIGIN_AIRPORT','ORIGIN']:
    if c in bts_df.columns:
        origin_id_col = c
        break
for c in ['DEST_AIRPORT_ID','DEST_ID','DEST_AIRPORT','DEST']:
    if c in bts_df.columns:
        dest_id_col = c
        break

print("Detected origin id column:", origin_id_col)
print("Detected dest id column  :", dest_id_col)

# Coerce to numeric (safe) and create canonical columns ORIGIN_AIRPORT_ID / DEST_AIRPORT_ID
if origin_id_col:
    bts_df['ORIGIN_AIRPORT_ID'] = pd.to_numeric(bts_df[origin_id_col], errors='coerce')
else:
    bts_df['ORIGIN_AIRPORT_ID'] = np.nan

if dest_id_col:
    bts_df['DEST_AIRPORT_ID'] = pd.to_numeric(bts_df[dest_id_col], errors='coerce')
else:
    bts_df['DEST_AIRPORT_ID'] = np.nan

# Remove any previously attempted origin_iata/dest_iata columns (they were empty/misleading)
if 'origin_iata' in bts_df.columns:
    del bts_df['origin_iata']
if 'dest_iata' in bts_df.columns:
    del bts_df['dest_iata']

# Quick checks
print("Unique ORIGIN_AIRPORT_ID (sample 10):", bts_df['ORIGIN_AIRPORT_ID'].dropna().unique()[:10])
print("Count non-null ORIGIN_AIRPORT_ID:", int(bts_df['ORIGIN_AIRPORT_ID'].notna().sum()), "/", len(bts_df))
print("Count non-null DEST_AIRPORT_ID  :", int(bts_df['DEST_AIRPORT_ID'].notna().sum()), "/", len(bts_df))

# If you plan to use mapping file later, keep id_to_iata building but do NOT rely on it for core pipeline
# (id_to_iata may or may not exist in the notebook; we do not override it here)
```

```python
# Cell 8
af_small = af[['iata','lat','lon']].dropna(subset=['iata']).copy()
af_small['iata'] = af_small['iata'].astype(str).str.upper()

# ensure origin_iata/dest_iata are strings or NaN
bts_df['origin_iata'] = bts_df['origin_iata'].astype(str).replace({'none':'', 'nan':''}).replace('', np.nan)
bts_df['dest_iata'] = bts_df['dest_iata'].astype(str).replace({'none':'', 'nan':''}).replace('', np.nan)

bts_df = bts_df.merge(af_small.rename(columns={'iata':'origin_iata','lat':'origin_lat','lon':'origin_lon'}), on='origin_iata', how='left')
bts_df = bts_df.merge(af_small.rename(columns={'iata':'dest_iata','lat':'dest_lat','lon':'dest_lon'}), on='dest_iata', how='left')

# convert scheduled times (HHMM) to minutes
def hhmm_to_minutes(x):
    try:
        if pd.isna(x): return None
        xi = int(float(x)) % 2400
        hh = xi // 100; mm = xi % 100
        return hh*60 + mm
    except Exception:
        return None

if 'CRS_DEP' in bts_df.columns:
    bts_df['CRS_DEP_MINS'] = bts_df['CRS_DEP'].apply(hhmm_to_minutes)
if 'CRS_ARR' in bts_df.columns:
    bts_df['CRS_ARR_MINS'] = bts_df['CRS_ARR'].apply(hhmm_to_minutes)

# build SCHED_DEP_DT and SCHED_ARR_DT best-effort
def make_sched_dt(row, mins_col):
    try:
        mins = row.get(mins_col)
        if pd.isna(mins): return pd.NaT
        year = int(row.get('YEAR', 2000))
        month = int(row.get('MONTH', 1))
        day = int(row.get('DAYOFMONTH', 1))
        hh = int(mins // 60); mm = int(mins % 60)
        return datetime(year, month, day, hh, mm)
    except Exception:
        return pd.NaT

bts_df['SCHED_DEP_DT'] = bts_df.apply(lambda r: make_sched_dt(r, 'CRS_DEP_MINS'), axis=1)
bts_df['SCHED_ARR_DT'] = bts_df.apply(lambda r: make_sched_dt(r, 'CRS_ARR_MINS'), axis=1)

# adjust arrival +1 day when arrival < dep (overnight)
mask = (~bts_df['SCHED_DEP_DT'].isna()) & (~bts_df['SCHED_ARR_DT'].isna()) & (bts_df['SCHED_ARR_DT'] < bts_df['SCHED_DEP_DT'])
bts_df.loc[mask, 'SCHED_ARR_DT'] = bts_df.loc[mask, 'SCHED_ARR_DT'] + pd.Timedelta(days=1)

print("Merged lat/lon and built scheduled datetimes.")
```

```python
# Cell 9
# ensure CANCELLED/DIVERTED exist
bts_df = bts_df[~(bts_df.get('CANCELLED',0).astype(int).eq(1) | bts_df.get('DIVERTED',0).astype(int).eq(1))]
bts_df['DIVERTED']  = pd.to_numeric(bts_df.get('DIVERTED', 0), errors='coerce').fillna(0).astype(int)

# drop rows with cancelled or diverted
bts_df = bts_df[~((bts_df['CANCELLED'] == 1) | (bts_df['DIVERTED'] == 1))].copy()
print("Rows after dropping cancelled/diverted:", len(bts_df))

# use ARR_DELAY for regression; fill from DEP_DELAY if needed (optional)
if 'ARR_DELAY' not in bts_df.columns and 'DEP_DELAY' in bts_df.columns:
    bts_df['ARR_DELAY'] = bts_df['DEP_DELAY']

bts_df['ARR_DELAY'] = pd.to_numeric(bts_df.get('ARR_DELAY'), errors='coerce')
# drop rows with missing ARR_DELAY (can't be used for supervised regression/classification)
bts_df = bts_df[~bts_df['ARR_DELAY'].isna()].copy()
print("Rows after dropping missing ARR_DELAY:", len(bts_df))

# regression target (minutes)
bts_df['DELAY_MINUTES'] = bts_df['ARR_DELAY']

# multiclass label: 0 on-time <=15, 1 minor 16-59, 2 major >=60
def mc_label(x):
    if x <= 15: return 0
    if x <= 59: return 1
    return 2
bts_df['DELAY_CLASS'] = bts_df['DELAY_MINUTES'].apply(mc_label).astype(int)

print("Class distribution:\n", bts_df['DELAY_CLASS'].value_counts())
```

```python
# Cell 10
# temporal features
bts_df['sched_dep_hour'] = bts_df['SCHED_DEP_DT'].dt.hour
bts_df['dayofweek'] = bts_df['SCHED_DEP_DT'].dt.dayofweek
bts_df['month'] = bts_df['SCHED_DEP_DT'].dt.month

# Top-K for origin/dest to avoid huge OHE
TOP_K = 30
top_orig = bts_df['ORIGIN_AIRPORT_ID'].value_counts().nlargest(TOP_K).index.tolist()
top_dest = bts_df['DEST_AIRPORT_ID'].value_counts().nlargest(TOP_K).index.tolist()

bts_df['origin_topK'] = bts_df['ORIGIN_AIRPORT_ID'].where(bts_df['ORIGIN_AIRPORT_ID'].isin(top_orig), other=-1)
bts_df['dest_topK']   = bts_df['DEST_AIRPORT_ID'].where(bts_df['DEST_AIRPORT_ID'].isin(top_dest), other=-1)

print("Top origin_topK (IDs):")
print(bts_df['origin_topK'].value_counts().head(10))
# candidate features (lightweight)
feature_candidates = [
    'YEAR','month','dayofweek','sched_dep_hour','DISTANCE','CRS_ELAPSED_TIME',
    'origin_topK','dest_topK','OP_UNIQUE_CARRIER'
]
feature_cols = [c for c in feature_candidates if c in bts_df.columns]

model_df = bts_df[feature_cols + ['DELAY_CLASS','DELAY_MINUTES']].copy()
print("Final model_df shape:", model_df.shape)
model_df.head()
```

```python
# Cell 11
# Prepare X and y for classification and regression (we'll build separate models)
X = model_df.drop(columns=['DELAY_CLASS','DELAY_MINUTES'])
y_cls = model_df['DELAY_CLASS']
y_reg = model_df['DELAY_MINUTES']

# split (stratify for classification)
X_train, X_test, y_train_cls, y_test_cls = train_test_split(X, y_cls, stratify=y_cls, test_size=0.2, random_state=RANDOM_STATE)
# For regression, align indexes (use same split indices)
X_train_reg = X_train.copy(); X_test_reg = X_test.copy()
y_train_reg = y_reg.loc[y_train_cls.index] if hasattr(y_reg,'loc') else y_reg.iloc[y_train_cls.index]
y_test_reg = y_reg.loc[y_test_cls.index] if hasattr(y_reg,'loc') else y_reg.iloc[y_test_cls.index]

# separate numeric and categorical columns
numeric_cols = X.select_dtypes(include=['number']).columns.tolist()
categorical_cols = X.select_dtypes(include=['object','category']).columns.tolist()

print("Numeric cols:", numeric_cols)
print("Categorical cols:", categorical_cols)

from sklearn.impute import SimpleImputer
numeric_transformer = Pipeline([('imputer', SimpleImputer(strategy='median')), ('scaler', StandardScaler())])
# sparse OHE
categorical_transformer = Pipeline([('imputer', SimpleImputer(strategy='constant', fill_value='MISSING')), 
                                    ('ohe', OneHotEncoder(handle_unknown='ignore', sparse_output =True))])

preprocessor = ColumnTransformer(transformers=[
    ('num', numeric_transformer, numeric_cols),
    ('cat', categorical_transformer, categorical_cols)
], remainder='drop', sparse_threshold=0.3)  # keep sparse when possible

# Fit-transform training data (classification pipeline)
X_train_pre = preprocessor.fit_transform(X_train)
X_test_pre = preprocessor.transform(X_test)

print("Preprocessed shapes (may be sparse):", type(X_train_pre), getattr(X_train_pre, "shape", None))

# Train small RandomForest classifier (balanced)
clf = RandomForestClassifier(n_estimators=50, max_depth=12, class_weight='balanced', random_state=RANDOM_STATE, n_jobs=-1)
clf.fit(X_train_pre, y_train_cls)

# Classification eval
y_pred = clf.predict(X_test_pre)
print("Classification report:")
print(classification_report(y_test_cls, y_pred))

# Train regressor on same features (RandomForestRegressor)
reg = RandomForestRegressor(n_estimators=50, max_depth=12, random_state=RANDOM_STATE, n_jobs=-1)
reg.fit(X_train_pre, y_train_reg)
y_pred_reg = reg.predict(X_test_pre)
print("Regression RMSE:", np.sqrt(mean_squared_error(y_test_reg, y_pred_reg)))
print("Regression R2:", r2_score(y_test_reg, y_pred_reg))

# Save artifacts
joblib.dump(preprocessor, PREPROCESSOR_PATH)
joblib.dump(clf, CLASSIFIER_PATH)
joblib.dump(reg, REGRESSOR_PATH)
print("Saved preprocessor to", PREPROCESSOR_PATH)
print("Saved models to", CLASSIFIER_PATH, "and", REGRESSOR_PATH)

# free memory
del X_train_pre, X_test_pre
gc.collect()
```

```python
# Cell 12
# Save processed model dataframe (small)
model_df.to_parquet(PROCESSED_OUTPUT, index=False)
print("Saved processed sample to:", PROCESSED_OUTPUT)

# quick EDA prints
print("Delay class counts:\n", model_df['DELAY_CLASS'].value_counts())
print("Top carriers:\n", bts_df['OP_UNIQUE_CARRIER'].value_counts().nlargest(10))
print("Top origin_iata mapped:\n", bts_df['origin_iata'].value_counts().nlargest(10))
```
